```python
# Import the library
import plotly.express as px
import pandas as pd

df = pd.read_csv("titanic passenger list.csv")

# Create the chart:
fig = px.parallel_coordinates(
    df, 
    color="survived", 
    labels={"pclass": "pclass","survived": "survived", "name": "name", "sex": "sex", "age": "age", 
           "sibsp": "sibsp","parch": "parch", "ticket": "ticket", "fare": "fare", "age": "age", 
           "cabin": "cabin","embarked": "embarked", "boat": "boat", "body": "body", "home.dest": "home.dest"},
    color_continuous_scale=px.colors.diverging.Tealrose,
    color_continuous_midpoint=2)

# Hide the color scale that is useless in this case
fig.update_layout(coloraxis_showscale=False)

# Show the plot
fig.show()
```

```python
fig.write_html("customGraph.html")
```
